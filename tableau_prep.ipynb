```python
#Tableau HW - Diana Kennen
#Dependencies
import pandas as pd
import numpy as np
import os
import csv
```

```python
#Reading in mtv genre information and converting it into dataframes
jan_path = "data/202101-citibike-tripdata.csv"
feb_path = "data/202102-citibike-tripdata.csv"
mar_path = "data/202103-citibike-tripdata.csv"
apr_path = "data/202104-citibike-tripdata.csv"
may_path = "data/202105-citibike-tripdata.csv"
jun_path = "data/202106-citibike-tripdata.csv"

```

```python
#Create dataframes
jan_data = pd.read_csv(jan_path, skipinitialspace = True, low_memory = False)
feb_data = pd.read_csv(feb_path, skipinitialspace = True, low_memory = False)
mar_data = pd.read_csv(mar_path, skipinitialspace = True, low_memory = False)
apr_data = pd.read_csv(apr_path, skipinitialspace = True, low_memory = False)
may_data = pd.read_csv(may_path, skipinitialspace = True, low_memory = False)
jun_data = pd.read_csv(jun_path, skipinitialspace = True, low_memory = False)

```
